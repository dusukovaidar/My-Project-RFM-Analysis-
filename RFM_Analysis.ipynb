Next, I would like to perform some actions with conversion_value, conversion_value_margin and handling_cost.

First of all, as I understood, the margin is the difference between a product or service's selling price and the cost of production.

Thus, we can calculate our Profit considering handling cost as follow:

If a customer makes a purchase, Profit will be equal to the difference between conversion_value_margin and handling_cost.

In the case of "Cancel" or "Return" cases, Profit will be equal to handling cost. So, no Profit or negative Profit.

```python
data.loc[data['conversion_name'] == 'Purchase', 'Profit'] = data['conversion_value_margin'] - data['handling_cost']
data.loc[data['conversion_name'] == 'Cancel', 'Profit'] = data['handling_cost']
data.loc[data['conversion_name'] == 'Return', 'Profit'] = data['handling_cost']
data.head()
```

RFM Analysis is a method used to determine customer value by looking at three dimensions:

1) Recency: when is the last time the user takes an action?

2) Frequency: how many times does the user take this action?

3) Monetary value: what is the sum of monetary value from this user throughout his/her lifetime?

One interesting issue is how identify every individual person purchasing a product.

First name and Last name can not give us the whole picture since they can be duplicated (different persons can have the same name) and even their combination can have multiple appearences.

The email column also not applicable, because one person usually have multiple emails. 

So, I decided to choose the phone as an indicator of individuality. Often one person have one phone number (I believe it). 

```python
# Calculating the Recency
data_Recency = data.groupby(by='phone', as_index = False)['conversion_date'].max()
data_Recency.columns = ['Phone', 'Last_purchase']
recent_date = data_Recency['Last_purchase'].max()
data_Recency['Recency'] = data_Recency['Last_purchase'].apply(lambda x: (recent_date - x).days)
data_Recency.head()
```

```python
# Calculating Frequency
data_Frequency = data.groupby(by=['phone'], as_index=False)['conversion_date'].count()
data_Frequency.columns = ['Phone', 'Frequency']
data_Frequency.head()
```

```python
# Calculating Monetary
data_Monetary = data.groupby(by='phone', as_index=False)['Profit'].sum()
data_Monetary.columns = ['Phone', 'Monetary']
data_Monetary.head(10)
```

```python
# Merging 3 parameters
data_RF = data_Recency.merge(data_Frequency, on='Phone')
data_RFM = data_RF.merge(data_Monetary, on='Phone').drop(columns='Last_purchase')
data_RFM.head()
```

```python
# Merging all three columns in one dataframe
data_RFM['R_rank'] = data_RFM['Recency'].rank(ascending=False)
data_RFM['F_rank'] = data_RFM['Frequency'].rank(ascending=True)
data_RFM['M_rank'] = data_RFM['Monetary'].rank(ascending=True)

data_RFM['R_rank_norm'] = (data_RFM['R_rank']/data_RFM['R_rank'].max())*100
data_RFM['F_rank_norm'] = (data_RFM['F_rank']/data_RFM['F_rank'].max())*100
data_RFM['M_rank_norm'] = (data_RFM['F_rank']/data_RFM['M_rank'].max())*100

data_RFM.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)

data_RFM.head()
```

```python
# Calculating RFM score
# Numbers 0.1 0.3 and 0.6 were chosen arbitrary, based on my own feelings
data_RFM['RFM_Score'] = 0.1 * data_RFM['R_rank_norm'] + 0.3 *	data_RFM['F_rank_norm'] + 0.6 * data_RFM['M_rank_norm']
data_RFM['RFM_Score'] *= 0.05
data_RFM = data_RFM.round(2)
data_RFM[['Phone', 'RFM_Score']].head(10)
```

```python
# Visualization
plt.hist(data_RFM['RFM_Score'])
plt.xlabel('RFM_Score')
plt.ylabel('Phone')
plt.show()
```

```python
# Rating Customers based upon the RFM score
data_RFM["Customer_segment"] = np.where(data_RFM['RFM_Score'] > 4.5, "Top Customers", (np.where(data_RFM['RFM_Score'] > 4, "High value Customer", (np.where(data_RFM['RFM_Score'] > 3,\
							"Medium Value Customer", np.where(data_RFM['RFM_Score'] > 1.6, 'Low Value Customers', 'Lost Customers'))))))
data_RFM[['Phone', 'RFM_Score', 'Customer_segment']].head()
```

```python
# Visualization
plt.pie(data_RFM.Customer_segment.value_counts(), labels=data_RFM.Customer_segment.value_counts().index, autopct='%.0f%%')
plt.show()
```
